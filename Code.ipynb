```python
import mlflow

# Set the run ID of the successful run you want to retrieve hyperparameters from
run_id = "6ecacdc3d3634e558fc26205577f11c5"

# Connect to the MLflow tracking server
mlflow.set_tracking_uri("sqlite:///mlruns.db")  # Set the URI to your MLflow tracking server

# Fetch the run details
run_info = mlflow.get_run(run_id)

# Fetch the run's parameters (hyperparameters)
params = run_info.data.params
print("Hyperparameters:")
for key, value in params.items():
    print(f"{key}: {value}")
```
